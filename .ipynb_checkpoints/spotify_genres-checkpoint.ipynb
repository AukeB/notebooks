## Functions

### Access token

```python
def get_spotify_access_token(client_id, client_secret):
    # Spotify API endpoint for token retrieval
    token_url = "https://accounts.spotify.com/api/token"

    # Encode client ID and client secret
    client_creds = f"{client_id}:{client_secret}"
    client_creds_b64 = base64.b64encode(client_creds.encode()).decode()

    # Request headers
    headers = {"Authorization": f"Basic {client_creds_b64}"}

    # Request body
    data = {
        "grant_type": "client_credentials",
        "scope": SCOPES,
    }

    try:
        # Send POST request to retrieve access token
        response = requests.post(token_url, headers=headers, data=data)
        response.raise_for_status()
        token_info = response.json()
        access_token = token_info["access_token"]
        logging.info(f"Succesfully obtain access token for user {SPOTIFY_USERNAME}")
        return access_token
    except requests.exceptions.RequestException as e:
        logging.error(f"Error retrieving access token: {e}")
        return None
```

### API

```python
def get_all_user_playlists(access_token: str, url: str) -> list[str]:
    """ """
    response = requests.get(
        url=url,
        headers={"Authorization": f"Bearer {access_token}"},
        params={"limit": 50},
    )

    # Obtain playlist dictionary.
    playlist_dict = response.json()

    # Obtain a list of the user's playlist names.
    playlist_names = {
        x.get("name"): x.get("id") for x in playlist_dict.get("items", {})
    }

    logging.info(f"Found {len(playlist_names)} playlists.")

    return playlist_names
```

```python
def get_playlist_information(playlist_id: str, access_token: str) -> Dict:
    # Spotify API endpoint for fetching playlist tracks
    playlist_tracks_url = f"https://api.spotify.com/v1/playlists/{playlist_id}/tracks"

    # Request headers
    headers = {"Authorization": f"Bearer {access_token}"}

    playlist_data = []

    # Initial request
    response = requests.get(playlist_tracks_url, headers=headers)
    response.raise_for_status()
    data = response.json()

    # Add tracks from initial request
    playlist_data.append(data)
    next_link: str = data["next"]

    # Check if there are more tracks
    while next_link:
        # Make subsequent requests for more tracks
        response = requests.get(next_link, headers=headers)
        response.raise_for_status()
        data = response.json()

        # Add tracks from subsequent requests
        playlist_data.append(data)
        next_link = data.get("next")

    logging.info(f"Successfully obtain all playlist data for playlist {playlist_id}")

    return playlist_data
```

```python
def get_artist_genres(access_token, artist_id):
    # Spotify API endpoint for fetching artist details
    artist_url = f"https://api.spotify.com/v1/artists/{artist_id}"

    # Request headers
    headers = {"Authorization": f"Bearer {access_token}"}

    try:
        # Send GET request to retrieve artist details
        response = requests.get(artist_url, headers=headers)
        response.raise_for_status()
        artist_data = response.json()
        return artist_data.get("genres", [])
    except requests.exceptions.RequestException as e:
        logging.error(f"Error retrieving artist genres: {e}")
        return []
```

```python
def update_artist_id_genre_mapping(
    df: pd.DataFrame,
    access_token: str,
    file_name: str = "artist_id_genre_mapping.json",
    save_path: str = "data/",
) -> None:
    """ """
    file_path = f"{save_path}{file_name}"
    artist_id_genre_dict = {}

    # New column to store genres
    df["artist_genres"] = None

    # Obtain all unique artist_ids from the playlist.
    artist_ids = set(
        value for dictionary in df["artist_ids"] for value in dictionary.values()
    )

    # Load Artist ID to genre mapping if it exists.
    try:
        existing_mapping = load_mapping_from_json(file_name)
    except:
        existing_mapping = {}

    # Obtain genres
    for artist_id in artist_ids:
        if artist_id not in existing_mapping.keys():
            artist_id_genre_dict[artist_id] = get_artist_genres(access_token, artist_id)

    # Update Artist ID genre .json file.
    export_mapping_to_json(mapping_dict=artist_id_genre_dict, file_name=file_name)
```

### Loading json mappings

```python
def load_mapping_from_json(
    file_name: str,
    save_path: str = "data/",
) -> Dict:
    """ """
    file_path = f"{save_path}{file_name}"

    with open(file_path, "r") as f:
        return json.load(f)


def export_mapping_to_json(
    mapping_dict: Dict,
    file_name: str,
    save_path: str = "data/",
) -> None:
    """ """
    file_path = f"{save_path}{file_name}"

    try:
        existing_mapping = load_mapping_from_json(file_name=file_name)
    except:
        existing_mapping = {}

    existing_mapping = dict(sorted(existing_mapping.items()))
    existing_mapping.update(mapping_dict)

    with open(file_path, "w") as f:
        json.dump(existing_mapping, f)
```

### Formatting

```python
def playlist_information_to_dataframe(playlist_information: List):
    # Create empty list which in the end we will convert to a Pandas DataFrame.
    data_for_dataframe: list = []
    track_information: list = []
    artist_name_id_mapping: dict[str, str] = {}

    # Extracting track information from the dictionary
    for element in playlist_information:
        try:
            track_information.extend(element["tracks"]["items"])
        except:
            track_information.extend(element["items"])

    # Extracting track names and artist names
    for track_info in track_information:
        track_name = track_info["track"]["name"]
        artist_names = [artist["name"] for artist in track_info["track"]["artists"]]
        artist_ids = [artist["id"] for artist in track_info["track"]["artists"]]

        artist_ids_dict = {
            artist_name: artist_id
            for artist_name, artist_id in zip(artist_names, artist_ids)
        }

        artist_name_id_mapping.update(artist_ids_dict)

        data_for_dataframe.append(
            {
                "track_name": track_name,
                "artist_names": artist_names,
                "artist_ids": artist_ids_dict,
            }
        )

    # Creating a DataFrame
    df = pd.DataFrame(data_for_dataframe)

    return df, artist_name_id_mapping
```

```python
def add_genres_to_dataframe(
    df: pd.DataFrame,
    file_name: str = "artist_id_genre_mapping.json",
) -> pd.DataFrame:
    """ """
    # Obtain the Artist ID/Genre mapping from the .json file.
    artist_id_genre_mapping = load_mapping_from_json(file_name=file_name)
    genre_column_new_values = []

    # Loop through dictionary.
    for artist_name_id_mapping in df["artist_ids"]:
        artist_id_genre_dict = {}
        for _, artist_id in artist_name_id_mapping.items():
            artist_id_genre_dict[artist_id] = artist_id_genre_mapping[artist_id]
        genre_column_new_values.append(artist_id_genre_dict)

    df["artist_genres"] = genre_column_new_values
    return df
```

### Analysis

```python
def count_genre_occurrences(dataframe):
    genre_list = []
    for index, artist_genre_dict in dataframe["artist_genres"].items():
        for artist_name, artist_genre in artist_genre_dict.items():
            genre_list += artist_genre

    df_temp = pd.DataFrame({"genres": genre_list})
    df_genre_counts = df_temp["genres"].value_counts().reset_index()
    df_genre_counts.columns = ["genre", "count"]

    return df_genre_counts
```

### Plotting

```python
def plot_top_genres(genre_counts):
    # Select the top 50 genres
    top_genres = genre_counts.head(50)

    # Create the bar plot
    plt.figure(figsize=(12, 8))
    plt.bar(top_genres["genre"], top_genres["count"], color="skyblue")

    # Rotate x-axis labels for readability
    plt.xticks(rotation=40, ha="right", fontsize=10)

    # Set labels and title
    plt.xlabel("Genre", fontsize=12)
    plt.ylabel("Count", fontsize=12)
    plt.title("Top Genres", fontsize=14)

    # Adjust layout
    plt.tight_layout()

    # Show plot
    plt.show()
```

```python
# Obtain access token.
access_token = get_spotify_access_token(CLIENT_ID, CLIENT_SECRET)

columns = ["track_name", "artist_names", "artist_ids", "artist_genres"]
df_main = pd.DataFrame(columns=columns)

# Obtain the user's playlists.
playlists: dict[str, str] = get_all_user_playlists(
    access_token, url=f"https://api.spotify.com/v1/users/{SPOTIFY_USERNAME}/playlists"
)

# Loop through all playlists.
for _, playlist_id in playlists.items():
    # Obtain all information for a single playlist.
    playlist_information = get_playlist_information(playlist_id, access_token)

    # Turn it into a DataFrame and obtain mapping for artist name/id as dictionary.
    df_playlist, artist_name_id_mapping = playlist_information_to_dataframe(
        playlist_information
    )

    # Export the artist name/ID mapping to a json file (append).
    export_mapping_to_json(
        mapping_dict=artist_name_id_mapping,
        file_name=file_name_artist_name_id_mapping,
    )

    # Obtain genres (saved to the artist id/genre mapping .json file)
    update_artist_id_genre_mapping(
        df_playlist, access_token, file_name=file_name_artist_id_genre_mapping
    )

    # Update playlist DataFrame.
    df_playlist = add_genres_to_dataframe(
        df=df_playlist, file_name=file_name_artist_id_genre_mapping
    )

    df_main = pd.concat([df_main, df_playlist], ignore_index=True)
```
